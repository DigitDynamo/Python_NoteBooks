```python
# Question 6 Visualization 1
# Scatter plot of Bill Length and Bill Depth
plt.figure(figsize=(10,7))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",hue="species",data=penguins)
plt.xlabel("Bill Lenght (mm)")
plt.ylabel("Bill Depth (mm)")
plt.title("Scatter Plot of Bill Length and Bill Depth")
plt.show()
```

```python
# Question 6 Visualization 2
# Distribution of Body Mass by Species as Boxplot
plt.figure(figsize=(10, 7))
sns.boxplot(x="species", y="body_mass_g", data=penguins)
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.title("Distribution of Body Mass by Species")
plt.show()
```

```python
# Question 6 Visualization 3
# Line plot of Average Body Mass over the years
plt.figure(figsize=(10, 7))
penguins.groupby("year")["body_mass_g"].mean().plot(marker="o")
plt.xlabel("Year")
plt.ylabel("Average Body Mass (g)")
plt.title("Average Body Mass Over Years")
plt.show()
```

```python
# Question 6 Visualization 4
# Pairwise scatter Plot
sns.set(style="ticks")
sns.pairplot(data=penguins, vars=["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"], hue="species")
plt.suptitle("Pairwise Scatter Plot of Numeric Variables")
plt.show()
```

```python
# Question 6 Visualization 5
# Bar plot of island counts
plt.figure(figsize=(10, 7))
penguins["island"].value_counts().plot(kind="bar")
plt.xlabel("Island")
plt.ylabel("Count")
plt.title("Number of Penguins on Each Island")
plt.show()
```
